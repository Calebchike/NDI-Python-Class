## NumPy arrays and array properties

`np` is the standard alias for the NumPy library. Below we create a Python
list and convert it into a NumPy array using the `array` function.

```python
import numpy as np

# *
sales = [0, 5, 155, 0, 518, 0, 1827, 616, 317, 325]
sales_array = np.array(sales)
sales_array
```

Check the type of the array. It's an `ndarray` (N-dimensional array).

```python
type(sales_array)
```

Key array attributes: number of dimensions, shape, size, and data type.

```python
sales_array.ndim
```

```python
sales_array.shape
```

```python
sales_array.size
```

```python
sales_array.dtype
```

To create a two-dimensional array, pass in nested lists of equal length.

```python
# *
sales = [[0, 5, 155, 0, 518], [0, 1827, 616, 317, 325]]
sales_array = np.array(sales)
sales_array
```

## Array creation

Useful functions to create arrays from scratch as an alternative to
converting base Python data types.

The `ones` function creates an array of ones with the given dimensions
(stored as floats by default).

```python
np.ones((10, 2))
```

The `zeros` function creates an array of zeros. Here we specify two rows,
five columns, and an integer data type.

```python
np.zeros((2, 5), dtype=int)
```

`arange` is analogous to base Python's `range` but automatically creates a
NumPy array.

```python
np.arange(10)
```

`linspace` creates an array of evenly spaced floats given a start, stop
(inclusive), and number of elements.

```python
np.linspace(0, 100, 5)
```

The `reshape` method changes an array into compatible dimensions.

```python
np.arange(1, 8, 2).reshape(2, 2)
```

## Random number generation

Create a random number generator with a seed so results are reproducible
across sessions. `rng` is the standard variable name.

```python
from numpy.random import default_rng

rng = default_rng(12345)
random_array = rng.random(10)
random_array
```

Drawing from a normal distribution with a given mean and standard deviation.

```python
rng = default_rng()
mean = 5
std = 1
rng.normal(mean, std, 10)
```

## Indexing and slicing arrays

Access elements by indexing (single element) or slicing (a series of
elements). Python is zero-indexed and stop points are not inclusive.

```python
# *
product_array = np.array(["fruits", "vegetables", "cereal", "dairy", "eggs",
                          "snacks", "beverages", "coffee", "tea", "spices"])
```

Indexing a one-dimensional array. Index 1 grabs the second element; index
-1 grabs the last element (negative indexing works from the end).

```python
product_array[1]
```

```python
product_array[-1]
```

Slicing: with no start point it defaults to 0; with no step size it
defaults to 1.

```python
product_array[:5]
```

```python
product_array[5::2]
```

```python
# *
product_array2D = product_array.reshape(2, 5)
```

Slicing a two-dimensional array requires coordinates: row then column.
Coffee is in the second row (index 1), third column (index 2).

```python
product_array2D[1, 2]
```

Grab all rows, but slice columns from the third column to the end.

```python
product_array2D[:, 2:]
```

Grab all columns, but only the second row onward.

```python
product_array2D[1:, :]
```

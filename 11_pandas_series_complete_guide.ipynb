## Series Basics

Series are Pandas data structures built on top of NumPy arrays. Almost every
piece of data in a Pandas data structure has a NumPy array hiding underneath
it. One differentiating factor between a NumPy array and a Pandas Series is
that we can give a Series a custom index as well as an optional name, which is
analogous to a column header or column title seen in other tools.

Series can be created from other data types like base Python lists or NumPy
arrays, but more often we import them from external sources like CSV or Excel
files or SQL databases. Two or more Series grouped together form a Pandas
DataFrame.

### Version note


````
import numpy as np
import pandas as pd
````

Create a Python list of integers, then use the Pandas `Series` function to
create a Series out of it. The optional `name` argument names the Series.
When we return the Series, in addition to the array of data, it is stored with
an index. By default this index is the zero-based range of integers the length
of the data. The name and the data type are also returned.

````
sales = [0, 5, 15, 50, 25, 100, 75, 35, 60, 90]
sales_series = pd.Series(sales, name="sales")
sales_series
````

### Key Series properties

- `.values` - the data array in the Series (a NumPy array)
- `.index` - the index array in the Series (a NumPy array)
- `.name` - the Series name, useful for accessing columns in a DataFrame
- `.dtype` - the data type of the elements in the values array

````
sales_series.values
````

````
sales_series.index
````

````
sales_series.name
````

````
sales_series.dtype
````

### Series demo

Start by creating an array and converting it into a Pandas Series. The left
column is the integer index; the right column is the Series values. Because
both are the same range, they share the same values, with data type int64.

````
arr = np.arange(5)
pd.Series(arr)
````

We can skip a step and pass the array directly. Giving the Series a name stores
the name below the values.

````
pd.Series(np.arange(5), name="Test Array")
````

Series must be one-dimensional. Trying to convert a two-dimensional array
raises an error: the data must be one-dimensional. For rows and columns, use a
Pandas DataFrame.

````
try:
    two_d = np.arange(6).reshape(3, 2)
    pd.Series(two_d)
except Exception as e:
    print(f"{type(e).__name__}: {e}")
````

Store the Series and inspect its attributes. Once we access `.values`, we are
working with a NumPy array, so we can call NumPy methods on it directly.

````
series = pd.Series(np.arange(5))
series.values
````

````
series.values.mean()
````

Pandas has many similar methods to NumPy. We do not have to access `.values`
to use an aggregation - Pandas has its own `.mean()` built in.

````
series.mean()
````

The index is a RangeIndex. We can assign a new index, and we can change the
name after creation by assigning to the `.name` attribute.

````
series.index
````

````
series.index = [10, 20, 30, 40, 50]
series
````

````
series.name = "special series"
series
````

````
series.dtype
````

## Pandas Data Types and Type Conversion

Pandas data types mostly expand on their base Python and NumPy equivalents.
Until about 2020 all data types in Pandas were NumPy data types, but Pandas 1.0
introduced new types to address issues that arose when working with NumPy
types, particularly around null/missing values.

**Numeric types** (focus on the top three NumPy types for now):

- **Boolean** - stored as 1 (True) or 0 (False). The mean of a Boolean column
  gives the proportion of True values; the sum counts them.
- **Integer** - default 64-bit, can be lowered (e.g. int8, int32) to save memory.
- **Float** - default 64-bit, used for decimal values.

**Object and text types** - `object` is the default for any text. It is very
flexible but space-inefficient. Pandas introduced dedicated string and
categorical types in version 1.0.

**Time series types** - datetime, timedelta, and period. An entire section is
dedicated to time series data later in the course.

### Type conversion with `.astype()`

Convert the data type of a Series with `.astype()`, provided the types are
compatible. Start with an integer Series.

````
int_series = pd.Series([0, 5, 15, 50, 25])
int_series
````

Integer to float - each integer becomes its float equivalent.

````
int_series.astype(float)
````

Integer to Boolean - zero becomes False; any non-zero value becomes True.

````
int_series.astype(bool)
````

Boolean aggregations: `.mean()` gives the percent of values not equal to zero,
and `.sum()` counts the True values.

````
int_series.astype(bool).mean()
````

````
int_series.astype(bool).sum()
````

Converting to object or string is straightforward for numbers.

````
int_series.astype(str)
````

Converting an integer Series to a datetime fails because there is not enough
information about the unit (hours, minutes, seconds), raising a value error.

````
try:
    int_series.astype("datetime64")
except Exception as e:
    print(f"{type(e).__name__}: {e}")
````

Numbers convert to strings easily, but the reverse is not always possible.
Creating a Series from strings stores it as object, and trying to convert
non-numeric text like "a" to an integer raises a value error. Likewise,
calling `.mean()` on string data raises a type error.

````
str_series = pd.Series(["a", "b", "c"])
str_series
````

````
try:
    str_series.astype(int)
except Exception as e:
    print(f"{type(e).__name__}: {e}")
````

````
try:
    str_series.mean()
except Exception as e:
    print(f"{type(e).__name__}: {e}")
````

### Why `.astype()` matters

Very often when we read in data, there are values that force a numeric column
to be read in as an object or a string. We need to fix those values and then
convert to a numeric data type. At this stage, focus on `bool`, `int64`,
`float64`, and `object`.

## The Series Index and Custom Indices

The index is one of the key distinguishing factors between a NumPy array and a
Pandas Series. It allows easy access to "rows" in a Series (and DataFrame).

By default the index is a zero-based, monotonically increasing range of
integers. This is usually preferable - it allows easy indexing and supports
familiar Python slicing.

````
sales_series = pd.Series([0, 5, 15, 50, 25, 100, 75, 35, 60, 90], name="sales")
````

Grab the third element by positional index, or slice positionally (stop point
excluded, matching standard Python behaviour).

````
sales_series[2]
````

````
sales_series[2:5]
````

### Custom indices

Pandas allows us to replace the default integer index with custom labels.
Pass a list to the `index` argument at creation, or assign to `.index` later.

````
items = ["coffee", "bananas", "tea", "coconut", "sugar"]
sales_series = pd.Series([100, 80, 155, 60, 40], index=items, name="sales")
sales_series
````

Accessing a single value by custom label works like a dictionary lookup.

````
sales_series["tea"]
````

Slicing with custom labels returns a **closed** (inclusive) slice - the stop
point is included, because the next label is unknown. Slicing from "bananas"
to "coconut" returns three values.

````
sales_series["bananas":"coconut"]
````

### Custom index demo: Days

````
my_series = pd.Series(range(5))
my_series
````

With the default integer index, access the value at position 3, and slice
positionally (stop excluded). A step size argument works too.

````
my_series[3]
````

````
my_series[1:3]
````

````
my_series[1::2]
````

Recreate the same values with a custom day-based index.

````
my_series = pd.Series(
    range(5),
    index=["Day 0", "Day 1", "Day 2", "Day 3", "Day 4"],
)
my_series
````

````
my_series["Day 1"]
````

````
my_series["Day 2"]
````

Slicing from "Day 2" to "Day 4" returns the last three values - the stop point
is inclusive. To get just Day 2 and Day 3, slice exactly that start and stop.

````
my_series["Day 2":"Day 4"]
````

````
my_series["Day 2":"Day 3"]
````

Custom step sizes work with labels too - grab every other element.

````
my_series["Day 0":"Day 4":2]
````

## The `.iloc` Accessor

`.iloc` is the preferred way to access values by their **positional** index,
regardless of what the index labels actually are. It works even when a Series
has a custom non-integer index, is more efficient than traditional Python
slicing, and the stop point is **exclusive** (matching standard Python).

````
my_series = pd.Series(
    [0, 1, 2, 3, 4],
    index=["Day 0", "Day 1", "Day 2", "Day 3", "Day 4"],
)
````

Pass a single integer to grab one row.

````
my_series.iloc[2]
````

````
my_series.iloc[0]
````

````
my_series.iloc[-1]
````

Pass a list of integers to grab specific, non-consecutive rows.

````
my_series.iloc[[1, 3, 4]]
````

Pass a slice to grab a range of rows (stop excluded).

````
my_series.iloc[1:4]
````

````
my_series.iloc[0:5]
````

## The `.loc` Accessor

`.loc` is the preferred way to access data via its **labels**. It is the most
commonly used data access method in day-to-day work because we often reference
columns by name and rows by meaningful index values. The stop point is
**inclusive**.

````
sales_series = pd.Series(
    [100, 80, 155, 60, 40],
    index=["coffee", "bananas", "tea", "coconut", "sugar"],
    name="sales",
)
````

Single label lookup.

````
sales_series.loc["tea"]
````

Inclusive slice from "bananas" to "coconut".

````
sales_series.loc["bananas":"coconut"]
````

List of labels.

````
sales_series.loc[["coffee", "sugar"]]
````

### `.loc` with an integer index

`.loc` works even when the indices are integers. With the default index it
behaves almost identically to `.iloc`, except the stop point is inclusive.

````
my_series = pd.Series(
    [0, 1, 2, 3, 4],
    index=["Day 0", "Day 1", "Day 2", "Day 3", "Day 4"],
)
my_series.loc["Day 2"]
````

Passing a label that does not exist raises a key error, just like a dictionary
lookup.

````
try:
    my_series.loc["Day 22"]
except Exception as e:
    print(f"{type(e).__name__}: {e}")
````

Label-based slice including the stop point.

````
my_series.loc["Day 1":"Day 3"]
````

### Caution: non-consecutive integer indices

`.loc` looks at label values, not positions. If the integer index is
non-consecutive (e.g. after dropping rows), slicing can return unexpected
results. Here the index is `[0, 2, 3, 100, 5]`.

````
demo = pd.Series([0, 1, 2, 3, 4], index=[0, 2, 3, 100, 5])
demo
````

Slicing from label 0 to label 3 returns the rows labelled 0, 2, 3 - not
positional rows 0 through 3.

````
demo.loc[0:3]
````

Slicing from label 0 to label 5 includes everything in between, including 100.

````
demo.loc[0:5]
````

Resetting the index restores a clean consecutive integer index, making
positional-style slicing intuitive again.

````
demo_reset = demo.reset_index(drop=True)
demo_reset
````

````
demo_reset.loc[0:3]
````

## Duplicate Index Values and Resetting the Index

It is possible to assign duplicate index values to a Series without error, but
this is strongly discouraged. If a label appears more than once, a single
lookup returns multiple rows, so you are no longer guaranteed one row per
lookup. Best practice is to maintain at least one unique row identifier.

````
dup_series = pd.Series(
    [1, 2, 3, 4, 5],
    index=["Day 0", "Day 0", "Day 0", "Day 2", "Day 2"],
)
dup_series
````

````
dup_series.index
````

Looking up "Day 0" returns three rows; "Day 2" returns two.

````
dup_series.loc["Day 0"]
````

````
dup_series.loc["Day 2"]
````

We can still slice across the whole Series, but pinpointing a single row
becomes inconvenient.

````
dup_series.loc["Day 0":"Day 2"]
````

### Resetting the index

`reset_index()` restores the default range-based integer index. By default,
Pandas does not destroy the old index - it moves it into a column, producing a
DataFrame. Passing `drop=True` discards the old index entirely and keeps the
result as a Series.

Default behaviour - old index becomes a column (result is a DataFrame).

````
dup_series.reset_index()
````

With `drop=True` - old index is discarded, result stays a Series with a clean
zero-based integer index.

````
clean_series = dup_series.reset_index(drop=True)
clean_series
````

Now lookups are predictable again.

````
clean_series.loc[2]
````

````
clean_series.iloc[2]
````

Assign the date array as the index.

````
oil_series.index = dates
oil_series
````

Because the index is no longer a consecutive integer index, use positional
slicing with `.iloc` to grab the first 10 and last 10 prices.

````
oil_series.iloc[:10].mean()
````

````
oil_series.iloc[-10:].mean()
````

Use label-based `.loc` to slice the date range from 1 Jan 2017 to 7 Jan 2017
(inclusive stop). This returns four prices.

````
slice_2017 = oil_series.loc["2017-01-01":"2017-01-07"]
slice_2017
````

Reset the index back to integers, dropping the old date index so the result
stays a Series.

````
slice_2017.reset_index(drop=True)
````

## Indexing: Putting It All Together

Method summary:

- **Direct indexing `series[x]`** - works but less efficient, and ambiguous
  when index labels are integers. Prefer `.loc` or `.iloc`.
- **`.iloc[x]`** - always positional, exclusive stop. Best for grabbing
  first/last N rows or when index labels are non-integer.
- **`.loc[x]`** - always label-based, inclusive stop. Preferred for most use
  cases, especially with DataFrames and named columns.

When the index is the default consecutive integer index, `.loc` and `.iloc`
return nearly identical results; the main practical difference is the inclusive
vs. exclusive stop point.

## Common Pitfalls to Avoid

- **Forgetting the inclusive stop with `.loc`** - label-based slices always
  include the stop point, differing from standard Python slicing.
- **Non-consecutive integer index** - after dropping rows, `.loc` slices by
  label value, not position. Reset the index to avoid surprises.
- **Duplicate index values** - cause multi-row returns on single lookups. Aim
  for a unique row identifier.
- **Incompatible type conversion** - converting strings like "a" to integers
  raises a ValueError. Clean your data before converting types.

`.iloc` is the preferred way to access values by their **positional** index,

slicing, and the stop point is **exclusive** (matching standard Python).

Pass a slice to grab a range of rows (stop excluded).

````
my_series.iloc[1:4]
````

````
my_series.iloc[0:5]
````

## The `.loc` Accessor

`.loc` is the preferred way to access data via its **labels**. It is the most
commonly used data access method in day-to-day work because we often reference
columns by name and rows by meaningful index values. The stop point is
**inclusive**.

````
sales_series = pd.Series(
    [100, 80, 155, 60, 40],
    index=["coffee", "bananas", "tea", "coconut", "sugar"],
    name="sales",
)
````

Single label lookup.

````
sales_series.loc["tea"]
````

Inclusive slice from "bananas" to "coconut".

List of labels.

````
sales_series.loc[["coffee", "sugar"]]
````

It is possible to assign duplicate index values to a Series without error, but

lookup returns multiple rows, so you are no longer guaranteed one row per

````
dup_series = pd.Series(
    [1, 2, 3, 4, 5],
    index=["Day 0", "Day 0", "Day 0", "Day 2", "Day 2"],
)
dup_series
````

````
dup_series.index
````

Looking up "Day 0" returns three rows; "Day 2" returns two.

````
dup_series.loc["Day 0"]
````

````
dup_series.loc["Day 2"]
````

We can still slice across the whole Series, but pinpointing a single row
becomes inconvenient.

````
dup_series.loc["Day 0":"Day 2"]
````

Default behaviour - old index becomes a column (result is a DataFrame).

````
dup_series.reset_index()
````

With `drop=True` - old index is discarded, result stays a Series with a clean

## Indexing: Putting It All Together

Method summary:

- **Direct indexing `series[x]`** - works but less efficient, and ambiguous
  when index labels are integers. Prefer `.loc` or `.iloc`.
- **`.iloc[x]`** - always positional, exclusive stop. Best for grabbing
  first/last N rows or when index labels are non-integer.
- **`.loc[x]`** - always label-based, inclusive stop. Preferred for most use
  cases, especially with DataFrames and named columns.

When the index is the default consecutive integer index, `.loc` and `.iloc`
return nearly identical results; the main practical difference is the inclusive
vs. exclusive stop point.

## Filtering Series and Logical Tests

Filtering Series is very similar to filtering NumPy arrays. We filter by
passing a logical test into the `.loc` accessor. Traditional Python indexing
also works, but as a best practice prefer `.loc`. A logical test returns a
Boolean array, and only the rows where that array is `True` are selected.

````
sales_series = pd.Series(
    [0, 5, 0, 155, 518],
    index=["water", "coffee", "juice", "tea", "sugar"],
    name="sales",
)
sales_series
````

Filter to rows where sales are greater than zero. The zeros get filtered out.

````
sales_series.loc[sales_series > 0]
````

Conditions can be combined. Best practice is to store the combined condition
in a `mask` variable. Here we look for sales greater than zero **and** an index
equal to "coffee", which returns a single row. Combining an index filter with
a value filter is one way to work around duplicate index issues.

````
mask = (sales_series > 0) & (sales_series.index == "coffee")
sales_series.loc[mask]
````

### Logical operators and their Pandas method equivalents

In most cases use the base Python operators. Each has an equivalent Pandas
method, which can be more convenient when chaining further operations onto the
resulting Boolean array.

- Equality `==` -> `.eq()`
- Not equal `!=` -> `.ne()`
- Less than or equal `<=` -> `.le()`
- Less than `<` -> `.lt()`
- Greater than or equal `>=` -> `.ge()`
- Greater than `>` -> `.gt()`
- Membership `in` / `not in` -> `.isin()` / `~ .isin()`

A logical test returns a Boolean Series: `True` where the condition is met,
`False` otherwise. Here only one value equals 5.

````
sales_series == 5
````

The equivalent Pandas method gives the same result, and lets us chain further
operations such as `.mean()` directly onto the Boolean Series.

````
sales_series.eq(5)
````

````
sales_series.eq(5).mean()
````

### Membership tests with `.isin()`

The base Python `in` operator does not work element-wise on a Series, so we
use the `.isin()` method to test for membership. Here we test whether each
index value is in the list `["coffee", "tea"]`.

````
item_series = pd.Series(
    [10, 12, 20, 30, 40],
    index=["coffee", "coffee", "tea", "coconut", "sugar"],
    name="sales",
)
item_series.index.isin(["coffee", "tea"])
````

To test for "not in", place a tilde `~` in front of the Boolean array. The
tilde inverts Boolean values completely - anything `True` becomes `False` and
vice versa.

````
~item_series.index.isin(["coffee", "tea"])
````

### Filtering demo

````
my_series = pd.Series(
    [0, 1, 2, 3, 4],
    index=["Day 0", "Day 1", "Day 2", "Day 3", "Day 4"],
)
````

A simple equality test returns a Boolean array.

````
my_series == 2
````

Test for inequality.

````
my_series != 2
````

Filter the Series by passing the logical test into `.loc`.

````
my_series.loc[my_series != 2]
````

Use `.isin()` to keep only rows whose value is 1 or 2.

````
my_series.loc[my_series.isin([1, 2])]
````

Invert the membership test with a tilde to get the complement (values 0, 3, 4).

````
my_series.loc[~my_series.isin([1, 2])]
````

A greater-than filter. To invert this with a tilde, wrap it in parentheses
first - otherwise the tilde applies to `my_series` before the comparison.

````
my_series.loc[my_series > 2]
````

````
my_series.loc[~(my_series > 2)]
````

Using the `.gt()` method instead, no outer parentheses are needed because the
method operates on the array before the tilde inverts the result. Both
approaches are equivalent.

````
my_series.loc[~my_series.gt(2)]
````

Combining conditions with OR. This mask matches values in [1, 2] **or** values
greater than 2, which covers almost all rows. Using `&` instead would return
no rows, since no value can be both in [1, 2] and greater than 2.

````
mask = (my_series.isin([1, 2])) | (my_series > 2)
my_series.loc[mask]
````

A traditional membership test fails: it tries to evaluate the truth value of
the whole Series rather than operating element-wise. This is why `.isin()` is
required.

````
try:
    my_series in [1, 2]
except Exception as e:
    print(f"{type(e).__name__}: {e}")
````

## Missing Data Representation in Pandas

Missing data is often represented by NumPy's `NaN` ("not a number"). NaN is

database where a value was absent.

````
sales_series = pd.Series([0, 5, 155, np.nan, 518], name="sales")
sales_series
````

NumPy's NaN coerces any numeric column to float. Without the NaN this would be
int64. Operations on NaN return NaN by default.

operation.

### Pandas' own NA type

Pandas released its own missing data type, `pd.NA`, in December 2020. It allows

## Identifying Missing Data

`.isna()` returns `True` where a value is missing and `False` otherwise. It

summarizing missing values convenient.

````
checklist = pd.Series(
    ["complete", np.nan, np.nan, np.nan, "complete"],
    name="checklist",
)
checklist.isna()
````

Summing the Boolean Series counts the missing values (True counts as 1).

````
checklist.isna().sum()
````

Taking the mean gives the proportion of missing values.

````
checklist.isna().mean()
````

`value_counts()` suppresses missing values by default; pass `dropna=False` to
include them alongside the other categories.

````
checklist.value_counts(dropna=False)
````

## Fixing Missing Data

Handling missing data is highly problem-specific. Two core methods are
`.dropna()` (remove missing rows) and `.fillna()` (replace missing values).
Always understand *why* data is missing before deciding how to handle it.

`.dropna()` removes rows with missing values. Note the resulting gaps in the
index - use `.reset_index(drop=True)` to restore a clean index.

````
checklist.dropna()
````

For categorical data, filling missing values with a new category (e.g.
"incomplete") is often the cleanest approach - it preserves the rows and the
missingness becomes its own analyzable category.

````
checklist.fillna("incomplete")
````

### Missing data demo

````
my_series = pd.Series([np.nan] * 5)
my_series
````

Count and proportion of missing values - here everything is missing.

````
my_series.isna().sum()
````

````
my_series.isna().mean()
````

NumPy NaN cannot be cast to integer - this raises an IntCastingNaNError.

````
try:
    my_series.astype(int)
except Exception as e:
    print(f"{type(e).__name__}: {e}")
````

A Series of `pd.NA` defaults to object, but can be cast to a nullable integer
without the casting error.

````
na_series = pd.Series([pd.NA] * 5).astype("Int64")
na_series
````

Detection methods still work on the nullable integer Series.

````
na_series.isna().sum()
````

With `dropna=False`, value_counts reports the missing values (still shown as
NaN here due to a current bug).

````
na_series.value_counts(dropna=False)
````

A mixed Series with two missing values.

````
my_series = pd.Series([0, pd.NA, pd.NA, 3, 4]).astype("Int64")
my_series
````

Fill missing values with 0.

````
my_series.fillna(0)
````

Fill missing values with the mean. The result becomes a float because the mean
is a float.

````
my_series.fillna(my_series.mean())
````

Drop missing values, then reset the index. `drop=True` keeps the result a
Series rather than pushing the old index into a DataFrame column.

````
my_series.dropna().reset_index(drop=True)
````

## Challenge: Missing Data

An email from Rachel. Subject: Erroneous Data.

> I noticed that two prices were incorrect, so I had them filled in with
> missing values. Can you confirm the number of missing values in the price
> column? Once you've done that, fill the prices in with a median of the oil
> price series.

Recreate the dated oil series with two missing prices.

Count the missing values.

Fill the missing values with the median of the series.

````
oil_series.fillna(oil_series.median())
````
